```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',
                y='sales',
                data=df,
                #size = 'salary'             sizing scplts with salary from small to high
                hue='level of education',   #hue provides color by column
                #palette = 'Dark2'           palette to create custom color
                #alpha=0.2                   for transparency
                style = 'level of education');    #should be categorical
            
plt.savefig('my_sns_scatterplot.png')
```

### Distribution Plots 

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(8,3),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
#do not use DISTPLOT
sns.set(style='ticks')
sns.displot(x='salary',
            data=df,
            bins=20,
            color='red',
            edgecolor='blue',
            linewidth=2,
            kde=True)

plt.savefig('my_histogram_distr.png',bbox_inches='tight');
```

```python
sns.histplot(x='salary',data=df)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['Age'])
```

```python
sample_ages
```

```python
sns.set(style='darkgrid')
sns.rugplot(data=sample_ages,x='Age',height=0.5);
```

```python
sns.set(style='darkgrid')
sns.displot(data = sample_ages,
            x = 'Age',
            color='blue',
            edgecolor='red',
            bins = 30,
            kde=True,
            rug=True);
```

```python
# Density acts as a propability of some event occuring i.e Prob of picking someone @Age of 40 = 0.010
sns.set(style='whitegrid')
sns.kdeplot(data=sample_ages,x='Age',clip=[0,100],bw_adjust=0.1,shade=True)  
```

### Categorical Plots - Statistics with Categories

```python
df.head(2)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.countplot(data=df,x='division') 
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division',palette='magma') #Kinda decieving for master's degree
plt.title('Number of Employees by Level of education and Division')
```

```python
df.pivot_table(index=['level of education','division'],values='salary',aggfunc='mean')
```

```python
plt.figure(figsize=(8,3),dpi=200)
sns.set(style='darkgrid')
sns.barplot(data=df,
            x='level of education',
            y='salary', 
            estimator=np.mean,
            ci='sd',
            hue='division')


plt.title('Average salary by Level of Education and SD')
plt.legend(bbox_to_anchor=(1.05,1))
plt.savefig('sns_barplot.png',bbox_inches='tight')
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
#BOXPLOT

#We expect students who have completed test prep,
#or those with parental level of education in (college ad more) to perfom better
#Results = High median and boxplot more tighter than those who haven't done.

plt.figure(figsize=(10,5),dpi=200)
sns.boxplot(data=df,
            y='parental level of education',
            x='math score',
            hue='test preparation course',
            palette='magma')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
#VIOLINPLOT

plt.figure(figsize=(10,5),dpi=200)
sns.violinplot(data=df,
            y='parental level of education',
            x='reading score',
            )
            #hue='test preparation course',
            #palette='Dark2',
            #split=True,
            #bw=0.08

#plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
#SWARMPLOT

plt.figure(figsize=(10,3),dpi=200)
sns.swarmplot(data=df,
              x='math score',  #you can use x only (optional)
              y='gender',
              hue='test preparation course',
              palette='Set2',
              size=3)

plt.legend(bbox_to_anchor=(-0.15,0.8))
plt.savefig('my_swarmplot.png',bbox_inches='tight')
```

```python
#BoxenPLOT

sns.boxenplot(x='math score',
              y='test preparation course',
              data=df,
              hue='gender');
```

```python
df.head()
```

```python
#JointPlot checking for relationship in continuous feautures and by using hue; you can also compare with categorical features

sns.jointplot(data=df,
              x='math score',
              y='reading score',
              hue='gender')
```

```python
sns.pairplot(data=df,hue='gender')
```

### Matrix Plots

```python
df = pd.read_csv('country_table.csv')
df
```

```python
df = df.set_index('Countries')
df
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),
                    lw=0.5,
                    annot=True,
                    cmap = 'viridis')   #colormapping
```

```python
#clustermap groups together similar feautures

plt.figure(figsize=(10,4),dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1),
                    lw=0.5,
                    annot=True,
                    cmap = 'viridis',
                    col_cluster=False)   #colormapping
```
