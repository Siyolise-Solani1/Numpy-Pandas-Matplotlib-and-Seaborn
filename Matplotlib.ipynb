### Figure Object

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
x = np.arange(10)
x
```

```python
y = 2*x
y
```

```python
fig = plt.figure()

#Large set of axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#small set of axes
axes2 = fig.add_axes([1,1,0.5,0.5])  #x&y startpoints and width& heights of axes
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,20)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')


plt.show()
```

### Figure Parameters

```python
fig = plt.figure(figsize=(4,2),dpi=120)  #dpi is like resolution

#Large set of axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('newfig.png',bbox_inches='tight')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][1].plot(x,y)
axes[1][0].plot(a,b);
```

```python
type(axes)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(4,3),dpi=120)

for ax1,ax2 in axes:
    ax1.plot(x,y)
    ax2.plot(a,b)
    
fig.suptitle('Figure Level',fontsize='16')
plt.tight_layout()
# or fig.subplots_adjust(wspace=0.6,hspace=0.5)

fig.savefig('new_subplots.png',bbox_inches ='tight')
```

### Matplotlib Styling - Legends & Visuals

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='X vs X')
ax.plot(x,x**2, label='X power 2')

ax.legend(loc=(1.1,0.5)) #locating the legend
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=3,ls='--',marker='*',ms=10,label='X vs X') #RGB HEX Code
ax.plot(x,x**2,color='#428715',lw=2,ls='-.',marker='s',ms=8,label='X power 2')

ax.legend(loc=(1.1,0.5)) #locating the legend
```
