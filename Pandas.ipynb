### Inner Merge

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bob','Claire','Tamia']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bob']})
```

```python
registrations
```

```python
logins
```

```python
#Inner
pd.merge(registrations,logins,how='inner',on='name')
```

```python
#Left
pd.merge(registrations,logins,how='left',on='name')
```

```python
#Right
pd.merge(logins,registrations,how='right',on='name')
```

```python
#Outer
pd.merge(registrations,logins,how='outer',on='name')
```

```python
#How many licenses of each product type did google purchase
```

```python
licenses = df7[['Company','Product','Licenses']]
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

```python
df7.head(2)
```

```python
pd.pivot_table(df7,index='Company',aggfunc='sum',values=['Licenses','Sale Price'])
```

```python
df7.groupby('Company').sum()[['Licenses','Sale Price']]  #better option
```

```python
pd.pivot_table(data=df7,
               index=['Account Manager','Contact','Product'],
               values=['Sale Price'],
               aggfunc='sum',   #applying agg function = sum into Sale Price
               fill_value=0,   #filling nan with zero
               margins=True)   #grandtotal
```
